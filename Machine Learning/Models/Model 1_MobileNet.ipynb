```python
TRAIN_DIR='/content/train'
VAL_DIR='/content/validation'
```

```python
train_datagen=ImageDataGenerator(1./255)
train_generator=train_datagen.flow_from_directory(
    TRAIN_DIR,
    target_size=(300,300),
    batch_size=64,
    class_mode='binary'
)
```
